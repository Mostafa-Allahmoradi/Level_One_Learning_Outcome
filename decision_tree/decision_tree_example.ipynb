Decision trees are used for classification and regression problems. Think of them as decision blocks in a flow chart representing the decision making process. 
    - They segregate data based on features into a flow or decision making branch 
    - A flow is uncovered that produces the best prediction 
    - If a branch is not needed or it's irelevant it's removed(pruned)
    - The decision tree depth can be configured using hyperparameters 

Random forest:
    - It is set or ensemble of decision trees 
    - Each tree is created from a different sample of rows 
    - Each tree makes its own prediction 
    - All the predictions are averaged into final result
