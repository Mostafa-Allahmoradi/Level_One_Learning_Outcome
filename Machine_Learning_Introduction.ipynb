#### Machine Learning
Machine learning is a subset of AI that applies mathematics to large datasets to find trends and patterns in a given dataset. For example, given a dataset of home features (age, location, and the number of bedrooms), what will be the selling price of a specific home?

This is done without implicitly programmed by a developer or software engineer.

Machine learning mappings uncovered between the inputs and the outputs are stored in a mathematical model.

Models are often depicted as a brain because they are meant to simulate human intelligence.



#### Supervised learning

Supervised learning is a type of machine learning where the machines are trained using labeled data

Labelled data is data that already contains the target value the machine needs to learn how to predict
